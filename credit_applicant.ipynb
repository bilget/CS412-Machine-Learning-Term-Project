**Data Cleaning**

Since the training data includes missing attributes and less relavent features we need to clean the data before training. 

**Handling Missing Attributes**

**desc_**: We dropped that column because nearly 95% of this column is empty and we tought it wouldn't give us much information about target label.

**annual_inc_jointâ€™ **: We dropped that column because nearly 95% of this column is empty.

**dti_jointâ€™ **: We dropped that column because nearly 95% of this column is empty too.

**title**: we converted the empty values in this coulmn to "Other". Since we didnt know for sure whether this person doesn't have a job our data attribute is just missing.

**emp_title**: We dropped that column because attributes are not unique. For same job position there are different entries for differnet person. This will decrease the benefit of this column in training.

**emp_length**: We filled the empty values in this column with "<1 year". We taught that these are the people that didn't have any job. We belived that handling them in one class will increase the accuracy of our training.

**application_type**: We examined this column and we found that there are two categories in this column and most of the people are in the category "Individual". Therefore, we filled the empty one with "Individual".

**dti**: We calculated the mean for this column. Then we filled empty ones with the mean value.

**inq_last_mnths**: We counted the unique data in this column. We found that "0.0" is the most occured one. Therefore we filled the empty values with "0.0".

**total_bal_il** : We dropped this column beause more than half of the rows are empty.
